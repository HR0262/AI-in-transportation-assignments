```python
plt.figure(figsize=(12, 12))
marginx=6
marginy=1
plt.xlim(-15 - 2, 15 + 5)
plt.ylim(-18 - marginy, 12 + marginy)


# 先画底层较粗较浅的线
nx.draw_networkx_edges(G, pos, edge_color='#f140a9', width=10, alpha=1)

# 再叠加画细一点颜色更深的线
nx.draw_networkx_edges(G, pos, edge_color='#E01695', width=2, alpha=1)

nodes = list(G.nodes())

# 目标节点列表
special_nodes = ['Uppsala C', 'Märsta', 'Bålsta', 'Södertälje centrum', 'Gnesta', 'Nynäshamn']
normal_nodes = [n for n in G.nodes() if n not in special_nodes]

# 第一层：绘制大粉色圆（特殊节点）
nx.draw_networkx_nodes(G, pos, nodelist=special_nodes, node_size=300, node_color='#f140a9')

# 第二层：在上层再绘制小白色圆（模拟套圈）
nx.draw_networkx_nodes(G, pos, nodelist=special_nodes, node_size=100, node_color='white')

# 普通节点：小白圆
nx.draw_networkx_nodes(G, pos, nodelist=normal_nodes, node_size=30, node_color='white')



labels = {node: node for node in G.nodes()}

left_label_nodes = ["Sundbyberg", "Spånga", "Barkarby", "Jakobsberg", "Kallhäll", "Kungsängen", "Bro", "Bålsta", "Södertälje hamn",
                    "Östertälje", "Rönninge", "Tumba", "Tullinge", "Flemingsberg", "Huddinge", "Stuvsta",
                    "Södertälje syd", "Järna", "Mölnbo", "Gnesta", "Rosersberg", "Märsta"]  # 标签显示在左边的节点
right_label_nodes = [n for n in G.nodes() if n not in left_label_nodes]

label_pos = {}
for node, (x, y) in pos.items():
    if node in left_label_nodes:
        label_pos[node] = (x - 0.5, y)
    else:
        label_pos[node] = (x + 0.5, y)

# 先绘制右边标签
right_labels = {n: n for n in right_label_nodes}
nx.draw_networkx_labels(G, {k: label_pos[k] for k in right_label_nodes},
                        labels=right_labels, font_size=15, horizontalalignment='left')

# 再绘制左边标签
left_labels = {n: n for n in left_label_nodes}
nx.draw_networkx_labels(G, {k: label_pos[k] for k in left_label_nodes},
                        labels=left_labels, font_size=15, horizontalalignment='right')


# 绘图
plt.title("Commuter Rail Network Topology (Initial Layout)")
plt.axis("off")
plt.tight_layout()
plt.show()
```
